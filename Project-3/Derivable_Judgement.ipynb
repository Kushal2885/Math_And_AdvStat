Part A – Theoretical Foundation 

1. What is Inferential Statistics?

 - Inferential statistics is a branch of statistics used to draw conclusions or make predictions about a population based on sample data. It helps researchers analyze patterns, relationships, and trends without studying every individual in the population.

2. What is Hypothesis Testing and its Components?

 - Hypothesis testing is a statistical method used to determine whether there is enough evidence to support a claim about a population. It involves comparing sample data against assumptions to make decisions.

    Components of Hypothesis Testing:
- Null Hypothesis (H₀): Assumes no effect or no relationship exists.
- Alternative Hypothesis (H₁): Assumes there is an effect or relationship.
- Significance Level (α): Probability threshold, usually 0.05.
- Test Statistic: Numerical value calculated from sample data.
-Decision Rule: Determines whether to reject or accept H₀.
3. Explain Confidence Interval and Critical Value

- A confidence interval is a range of values used to estimate a population parameter with a certain confidence level, such as 95%. It indicates the reliability of the estimate.

- A critical value is a threshold value used in hypothesis testing to decide whether the null hypothesis should be rejected. It depends on the significance level and the type of statistical test.

4. Define p-value

- The p-value is the probability of obtaining results as extreme as the observed data, assuming the null hypothesis is true. A small p-value (typically less than 0.05) indicates strong evidence against the null hypothesis.

5. Differentiate Type I and Type II Errors
    Type of Error	Meaning
        - Type I Error	Rejecting the null hypothesis when it is actually true (False Positive).
        - Type II Error	Accepting the null hypothesis when it is actually false (False Negative).
6. Brief Descriptions of z-test, t-test, chi-square test, and ANOVA test

- z-test
    A z-test is used to compare sample and population means when the population variance is known and the sample size is large.

- t-test
    A t-test is used to compare the means of two groups when the population variance is unknown and the sample size is small.

- Chi-square Test
    A chi-square test is used to determine whether there is a significant association between categorical variables.

- ANOVA Test
    ANOVA (Analysis of Variance) is used to compare the means of three or more groups to check whether at least one group mean differs significantly.

7. What is Covariance?

- Covariance measures the direction of the relationship between two variables. A positive covariance indicates variables move in the same direction, while a negative covariance indicates they move in opposite directions.

8. What is Correlation?

- Correlation measures both the strength and direction of the relationship between two variables. Its value ranges from -1 to +1, where:
    +1 indicates perfect positive correlation,
    -1 indicates perfect negative correlation,
    0 indicates no correlation.

```python
# Part B
import pandas as pd
import numpy as np

from scipy import stats
```

```python
df = pd.read_csv("health_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Q1

print("Hypothesis 1")
print("H0: Smoking has no effect on Diabetes prevalence")
print("H1: Smoking affects Diabetes prevalence")

print("\nHypothesis 2")
print("H0: Exercise frequency has no effect on BMI")
print("H1: Exercise frequency affects BMI")
```

```python
# Q2

#age

confidence = 0.95
mean_age = np.mean(df['age'])
std_error_age = stats.sem(df['age'])
ci_age = stats.t.interval(confidence,len(df['age']) -1, loc=mean_age, scale=std_error_age)
print("Mean Age:", mean_age)
print("95% Confidence Interval for Age:", ci_age)

#weight

mean_weight = np.mean(df['weight'])
std_error_weight = stats.sem(df['weight'])
ci_weight = stats.t.interval(confidence, len(df['weight'])-1, loc=mean_weight, scale=std_error_weight)
print("Mean Weight:", mean_weight)
print("95% Confidence Interval for Weight:", ci_weight)

```

```python
#Q3

alpha = 0.05
sample = df['bmi']
sample_mean = np.mean(sample)
population_mean = 25
sample_std = np.std(sample, ddof=1)
n = len(sample)

t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=n-1))
critical_value = stats.t.ppf(1 - alpha/2, df=n-1)

print("T-Statistic :", t_statistic)
print("P-Value :", p_value)
print("Critical Value :", critical_value)
```

```python
# Q4

smokers = df[df['smoking_status'] == 'Smoker']['bmi']
non_smokers = df[df['smoking_status'] == 'Non-Smoker']['bmi']
t_statistic, p_value = stats.ttest_ind(smokers, non_smokers)

print("T-Statistic :", t_statistic)
print("P-Value :", p_value)
```

```python
#Q5

contingency_table = pd.crosstab(df['smoking_status'], df['diabetes'])
print(contingency_table)

chi, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-Square Statistic :", chi)
print("P-Value :", p_value)
print("Degrees of Freedom :", dof)
```

```python
#Q6

group1 = df[df['age_group'] == '18-25']['bmi']

group2 = df[df['age_group'] == '26-35']['bmi']

group3 = df[df['age_group'] == '36-45']['bmi']

group4 = df[df['age_group'] == '46-60']['bmi']

group5 = df[df['age_group'] == '60+']['bmi']

f_statistic, p_value = stats.f_oneway(group1, group2, group3, group4, group5)
print("F-Statistic :", f_statistic)
print("P-Value :", p_value)
```

```python
#Q7

covariance = np.cov(df['age'], df['bmi'])[0, 1]
correlation = np.corrcoef(df['age'], df['bmi'])[0, 1]

print("Covariance :", covariance)
print("Correlation :", correlation)
```

```python
#Q8

print("Conclusion")
print("----------------------------------")
if p_value < 0.05:
    print("The statistical test is significant.")
    print("Reject Null Hypothesis (H0).")
else:
    print("The statistical test is not significant.")
    print("Accept Null Hypothesis (H0).")
print("----------------------------------")
print("The analysis was performed using inferential statistics techniques.")
print("Tests like T-Test, Chi-Square Test, ANOVA, Covariance, and Correlation were used to analyze the dataset.")
```
